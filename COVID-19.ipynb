```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("latest Covid-19 India Status1 (1).csv")
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns=['State','Total_Cases','Active','Deaths','Active_Ratio','Death_Ratio']
```

```python
df.head()
```

```python
df.sort_values(by='Active_Ratio',ascending=False)
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active_Ratio'],color='red',marker='o')
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State',fontsize=15)
plt.ylabel('Active_Ratio',fontsize=15)
plt.title('Line Plot For Active Ratio',fontsize=15)
plt.show()
```

```python
Total_cases_state=df.groupby(['State'])['Total_Cases'].max().sort_values(ascending=False).reset_index()
```

```python
Total_cases_state
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=75)
sns.barplot(x='State',y='Total_Cases',data=Total_cases_state)
plt.title('Total Cases in Each State',fontsize=20)
plt.xlabel('state',fontsize=30)
plt.ylabel('Total Cases',fontsize=30)
plt.xticks(rotation=90,fontsize=11)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=75)
sns.barplot(x='State',y='Total_Cases',data=Total_cases_state.head())
plt.title('Total Cases in Top 5 states',fontsize=15)
plt.xlabel('state',fontsize=30)
plt.ylabel('Total Cases',fontsize=30)
plt.xticks(rotation=90,fontsize=11)
plt.show()
```

```python
plt.plot(df['State'],df['Active_Ratio'],color='red',marker='o',label='Active Cases')
plt.plot(df['State'],df['Death_Ratio'],color='blue',marker='*',label='Death')
plt.legend()
plt.xticks(rotation=90,fontsize=10)
plt.show()
```

```python
"pie chart of active cases in top 10 states in India"
```

```python
plt.figure(figsize=(8,8))
plt.pie(x="Total_Cases",data=Total_cases_state.head(10),labels='State',autopct='%.0f%%')
plt.title('Active Cases in Top 10 States in India',fontsize=24)
plt.show()
```

# RECOVERED CASES 

```python
df['Recovered_cases']=df['Total_Cases']-(df['Active']+df['Deaths'])
```

```python
df
```

```python
rec_cases=df.groupby(['State'])['Recovered_cases'].max().sort_values(ascending=False).reset_index()
```

```python
rec_cases
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=75)
sns.barplot(x='State',y='Recovered_cases',data=rec_cases)
plt.title('Recovered Cases in Each State',fontsize=20)
plt.xlabel('state',fontsize=30)
plt.ylabel('Recovered Cases',fontsize=30)
plt.xticks(rotation=90,fontsize=11)
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
plt.plot(df['State'],df['Total_Cases'],color='red',marker='o',label='Total_cases')
plt.plot(df['State'],df['Recovered_cases'],color='yellow',marker='s',label='rec_cases')
plt.plot(df['State'],df['Deaths'],color='black',marker='*',label='deaths')
plt.xticks(rotation=90,fontsize=10)
plt.legend()
plt.show()
```
